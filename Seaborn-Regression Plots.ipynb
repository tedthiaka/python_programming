## Implot
Linear Model Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

tips = sns.load_dataset('tips')
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

## Markers

```python
g = sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
g = sns.lmplot(x='total_bill',y='tip',row='sex',col='time',data=tips)
```

```python
g = sns.lmplot(x='total_bill',y='tip',row='sex',col='time',data=tips,markers=['x'],scatter_kws={'s':100})
```
