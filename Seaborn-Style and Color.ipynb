```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

tips = sns.load_dataset('tips')
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

```python
sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
sns.countplot(x='sex',data=tips,palette='deep')
```
