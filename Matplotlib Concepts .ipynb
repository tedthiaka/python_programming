## Basic Matplotlib Commands

```python
plt.plot(x,y,'r')#'r' is the colour red
```

```python
plt.plot(x,y,'y')#'r' is the colour red
plt.xlabel('X axis title here')
```

```python
plt.plot(x,y,'g')#'r' is the colour red
plt.xlabel('X axis title here')
plt.ylabel('Y axis title here')
```

```python
plt.plot(x,y,'b')#'r' is the colour red
plt.xlabel('X axis title here')
plt.ylabel('Y axis title here')
plt.title('String Title here')
#plt.show()
```

```python
plt.plot(x,y,'o')#'o' gives dots
plt.xlabel('X axis title here')
plt.ylabel('Y axis title here')
plt.title('String Title here')
#plt.show()
```

## Matplotlib Object Oriented Method

## Introduction

```python
#1st Create a Figure(Empty Canvas)
fig = plt.figure()
#Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.9,0.9])#left,bottom,width,height

#Plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel('Set X Label')#Notice the use of set_ to begin methods 
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
```

```python
#1st Create a Figure(Empty Canvas)
fig = plt.figure()

#Add set of axes to figure
axes1 = fig.add_axes([0.1,0.1,0.9,0.9])#main axis
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])#inset axis
#Plot on that set of axes

#Larger Figure Axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('Set X Label')#Notice the use of set_ to begin methods 
axes1.set_ylabel('Set Y Label')
axes1.set_title('Axes 1')

#Larger Figure Axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('Set X Label')#Notice the use of set_ to begin methods 
axes2.set_ylabel('Set Y Label')
axes2.set_title('Axes 2')
```

```python
#1st Create a Figure(Empty Canvas)
fig = plt.figure()

#Add set of axes to figure
axes1 = fig.add_axes([0.1,0.1,0.9,0.9])#main axis
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])#inset axis
#Plot on that set of axes

#Larger Figure Axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('Set X Label')#Notice the use of set_ to begin methods 
axes1.set_ylabel('Set Y Label')
axes1.set_title('Axes 1')

#Larger Figure Axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('Set X Label')#Notice the use of set_ to begin methods 
axes2.set_ylabel('Set Y Label')
axes2.set_title('Axes 2')
```
