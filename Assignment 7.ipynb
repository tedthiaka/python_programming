```python
fhandle = open("mbox.txt","r")
email = dict()
for a in fhandle:
    if a.startswith("From:") or a.startswith("To:"):
        b = a.split()
        email[b[1]] = email.get(b[1],0) + 1
print(email)
print()
for c,d in email.items():
        print("Email: ",c)
        print("Occurences: ",d)
        print("")
```
