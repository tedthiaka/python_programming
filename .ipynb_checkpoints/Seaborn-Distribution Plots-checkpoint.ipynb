# Distribution Plots
* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

```python
import seaborn as sns
%matplotlib inline
```

# Data

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(4)
```

## Distplot

Distribution of an unvariated set of observations.

```python
sns.distplot(tips['total_bill'])
```

To remove the kde layer and just have the histogram use:

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## Jointplot
Match up two distplots for bivariate data. You can choose what parameter to compare

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

## Pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='time',palette='coolwarm')
```

```python
sns.pairplot(tips,hue='smoker',palette='coolwarm')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.pairplot(tips,hue='day',palette='coolwarm')
```

## rugplot

```python
sns.rugplot(tips['total_bill'])
```

## kdeplot

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```
