# Categorical Data Plots
* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

## Barplot and Countplot

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
import numpy as np
sns.set_style('whitegrid')
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.mean)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.sum)
```

## Countplot

```python
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='smoker',data=tips)
sns.despine()
```

```python
sns.countplot(x='day',data=tips)
sns.despine()
```

```python
sns.countplot(x='time',data=tips)
sns.despine()
```

## boxplot and violinplot

```python
sns.set_context('paper',font_scale=1)
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

### Orientation**

```python
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

### Hue

```python
sns.boxplot(x='day',y='total_bill',hue='time',data=tips,palette='coolwarm')
```

```python
sns.boxplot(x='day',y='total_bill',hue='day',data=tips,palette='coolwarm')
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,palette='coolwarm')
```

```python
sns.boxplot(x='smoker',y='tip',hue='time',data=tips,palette='coolwarm')
```

```python
sns.boxplot(x='day',y='total_bill',hue='sex',data=tips,palette='coolwarm')
```

## Violinplot

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips, palette = 'rainbow')
```

### Hue

```python
sns.violinplot(x = 'day',y = 'total_bill', hue = 'time', data = tips, palette = 'RdBu')
```

```python
sns.violinplot(x = 'day',y = 'total_bill', hue = 'sex', data = tips, palette = 'Set1')
```

```python
sns.violinplot(x = 'day',y = 'total_bill', hue = 'day', data = tips, palette = 'Set1')
```

### Split

```python
sns.violinplot(x = 'day',y = 'total_bill', hue = 'sex', data = tips, palette = 'Set1',split=True)
```

```python
sns.violinplot(x = 'day',y = 'total_bill', hue = 'smoker', data = tips, palette = 'Set1',split=True)
```

```python
sns.violinplot(x = 'day',y = 'total_bill', hue = 'time', data = tips, palette = 'Set1',split=True)
```

## Stripplot and swarmplot

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x = 'day',y = 'total_bill', hue = 'time', data = tips, palette = 'Set1',split=True)
```

### Combining violinplot and swarmplot

```python
sns.violinplot(x = 'day',y = 'total_bill', data = tips, palette = 'Set1',split=True)
sns.swarmplot(x = 'day',y = 'total_bill',  data = tips, color='black',split=True,size=3)
```

### You can combine swarmplot and splitplot with boxplot or violinplots

## Factorplot
Takes a "kind" paramter to adjust the plot type

```python
sns.factorplot(x='sex',y='total_bill',data=tips, kind='box')
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips, kind='violin')
```
