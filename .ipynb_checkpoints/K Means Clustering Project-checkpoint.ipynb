# K means Cluster Creation

```python
from sklearn.cluster import KMeans
```

Create an instnace of a KMeans model with 2 clusters.**

```python
kmeans = KMeans(n_clusters=2)
```

Fit the model to all the data except the Private label

```python
kmeans.fit(df.drop('Private',axis=1))
```

What are the cluster center vectors

```python
kmeans.cluster_centers_
```

## Evaluation
Since we are using clustering, we won't always have the luxury of being provided with labels for the clusters
But since this is an exercise, we know the labels

Create a new column for df called "Cluster" which is 1 for a Private school and 0 for "Public"

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster']= df['Private'].apply(converter)
```

```python
df.head()
```

Create a confusion matrix and a classification report to see how well the KMeans clustering worked without being given any labels

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```
