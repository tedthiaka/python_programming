# Grids
Allows you to map plot types to rows and columns and helps you create similar plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#Just the grid
sns.PairGrid(iris)
```

```python
#Map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter)
```
