# Examples

Using two arrays

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

## Basic Matplotlib Commands

```python
plt.plot(x,y,'r')#'r' is the colour red
```

```python
plt.plot(x,y,'y')#'r' is the colour red
plt.xlabel('X axis title here')
```

```python
plt.plot(x,y,'g')#'r' is the colour red
plt.xlabel('X axis title here')
plt.ylabel('Y axis title here')
```

```python
plt.plot(x,y,'b')#'r' is the colour red
plt.xlabel('X axis title here')
plt.ylabel('Y axis title here')
plt.title('String Title here')
#plt.show()
```

```python
plt.plot(x,y,'o')#'o' gives dots
plt.xlabel('X axis title here')
plt.ylabel('Y axis title here')
plt.title('String Title here')
#plt.show()
```

## Multiple plots on the same canvas

```python
#plt.sublot(nrows,ncols,plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')#More colour optins later
plt.subplot(1,2,2)
plt.plot(y,x,'g*-');
```

```python
#plt.sublot(nrows,ncols,plot_number)
plt.subplot(2,1,1)
plt.plot(x,y,'r--')#More colour optins later
plt.subplot(2,1,2)
plt.plot(y,x,'g*-');
```
